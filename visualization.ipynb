## Finish pine saplings dataset

### Raw

```julia
finpines = wload(datadir("finpines-raw.bson"))

scatter(finpines[:x], finpines[:y], size=(300, 300), label=nothing, aspect_ratio=:equal)
```

### Processed (with different number of grids)

```julia
ps = map([16, 32, 64]) do ngrid
    finpines = wload(datadir("finpines-$ngrid.bson"))
    data_counts = finpines[:data_counts]
    heatmap(reshape(data_counts, ngrid, ngrid), aspect_ratio=:equal, title="$ngrid x $ngrid")
end
plot(ps..., layout=@layout([a b c]), size=(300 * 3, 300))
```
